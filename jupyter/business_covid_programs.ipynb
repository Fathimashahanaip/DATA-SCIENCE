```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("business_covid.csv")
plt.figure(figsize=(10,3))
x=data["value"]
plt.hist(x,bins=20,color="green")
plt.title("value in cm")
plt.xlabel("value")
plt.ylabel("count")
plt.show
```
